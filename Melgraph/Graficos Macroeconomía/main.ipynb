```python
import pandas as pd
import matplotlib.pyplot as plt 

# Ruta al archivo Excel
excel_path = r"C:\Users\MELGRAPH01\Project-Python\Melgraph\Graficos Macroeconomía\Tabla - Tasa de Desempleo.xlsx"

# Leer el archivo Excel
df = pd.read_excel(excel_path)

# Eliminar columnas que están completamente vacías
df_cleaned = df.dropna(axis=1, how='all')

# Seleccionar solo las columnas deseadas
columns_of_interest = ['Año', 'República Dominicana', 'Chile', 'Costa Rica', 'Panamá', 'Uruguay', 'Perú']
df_filtered = df_cleaned[columns_of_interest]

df = df_filtered

# Redondear las columnas numéricas a 2 decimales
df.iloc[:, 1:] = df.iloc[:, 1:].round(2)

# Mostrar las primeras filas del DataFrame filtrado
print(df)

# Obtener valores mínimo y máximo por país
min_values = {}
max_values = {}

for column in df.columns[1:]:
    min_value = df[column].min()
    # Encontrar el año correspondiente al valor mínimo
    min_year = df.loc[df[column] == min_value, 'Año'].values[0]
    min_values[column] = (min_value, min_year)
    
    max_value = df[column].max()
    # Encontrar el año correspondiente al valor máximo
    max_year = df.loc[df[column] == max_value, 'Año'].values[0]
    max_values[column] = (max_value, max_year)

# Mostrar los resultados
print("\nValores mínimos por país (País: valor, Año):")
for country, (value, year) in min_values.items():
    print(f"{country}: {value} (Año: {year})")

print("\nValores máximos por país (País: valor, Año):")
for country, (value, year) in max_values.items():
    print(f"{country}: {value} (Año: {year})")    


```

```python
import pandas as pd
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import os

# Crear la aplicación Dash
app = Dash(__name__)

# Datos de tasas de desempleo
data = {
    'Año': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'República Dominicana': [8.26, 7.33, 7.08, 5.51, 5.66, 6.17, 5.83, 7.38, 5.29, 5.31],
    'Chile': [6.42, 6.39, 6.65, 6.93, 7.33, 7.22, 10.57, 9.11, 7.80, 8.64],
    'Costa Rica': [9.62, 9.61, 9.54, 9.07, 10.26, 11.76, 19.61, 16.43, 12.22, 8.89],
    'Panamá': [5.06, 5.29, 5.77, 6.44, 6.24, 7.42, 10.45, 11.84, 10.34, 7.70],
    'Uruguay': [6.55, 7.49, 7.84, 7.89, 8.34, 8.73, 10.33, 9.30, 7.88, 8.35],
    'Perú': [6.00, 6.43, 6.71, 6.87, 6.73, 6.58, 12.79, 11.31, 7.74, 6.86],
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico sin los mínimos y máximos inicialmente
def create_figure(show_min_max=False):
    fig = go.Figure()

    # Añadir trazas para cada país
    for country in df.columns[1:]:
        fig.add_trace(go.Scatter(
            x=df['Año'],
            y=df[country],
            mode='lines+markers',
            name=country,
            marker=dict(size=8),
            line=dict(width=2),
            opacity=0.8,
            hoverinfo='text',
            hovertext=df[country].apply(lambda x: f'{country}: {x:.2f}%')
        ))

    # Añadir valores máximos y mínimos si es necesario
    if show_min_max:
        min_values = {
            'República Dominicana': (5.29, 2022),
            'Chile': (6.39, 2015),
            'Costa Rica': (8.89, 2023),
            'Panamá': (5.06, 2014),
            'Uruguay': (6.55, 2014),
            'Perú': (6.00, 2014)
        }

        max_values = {
            'República Dominicana': (8.26, 2014),
            'Chile': (10.57, 2020),
            'Costa Rica': (19.61, 2020),
            'Panamá': (11.84, 2021),
            'Uruguay': (10.33, 2020),
            'Perú': (12.79, 2020)
        }

        # Añadir anotaciones para los mínimos
        for country, (min_value, min_year) in min_values.items():
            fig.add_trace(go.Scatter(
                x=[min_year],
                y=[min_value - 0.5],  # Ajustar la posición hacia abajo
                mode='markers+text',
                name=f'Mín: {country}',
                marker=dict(color='black', size=10, symbol='x'),
                text=[f'Mín: {min_value}'],
                textposition="bottom center",
                hoverinfo='none'
            ))

        # Añadir anotaciones para los máximos
        for country, (max_value, max_year) in max_values.items():
            fig.add_trace(go.Scatter(
                x=[max_year],
                y=[max_value + 0.5],  # Ajustar la posición hacia arriba
                mode='markers+text',
                name=f'Máx: {country}',
                marker=dict(color='red', size=10, symbol='x'),
                text=[f'Máx: {max_value}'],
                textposition="bottom center",
                hoverinfo='none'
            ))

    # Configurar el diseño
    fig.update_layout(
        title='Tasas de Desempleo por País (2014-2023)',
        xaxis_title='Año',
        yaxis_title='Tasa de Desempleo (%)',
        legend_title='País',
        hovermode='x unified',
        template='plotly_white',
        margin=dict(l=40, r=40, t=40, b=40),
        height=600,
    )

    return fig

# Layout de la aplicación
app.layout = html.Div([
    html.H1("Análisis de Desempleo en América Latina"),
    dcc.Graph(id='graph', figure=create_figure()),
    html.Button("Mostrar Mínimos y Máximos", id='toggle-button', n_clicks=0),
])

# Callback para actualizar el gráfico
@app.callback(
    Output('graph', 'figure'),
    Input('toggle-button', 'n_clicks')
)
def update_graph(n_clicks):
    # Determinar si mostrar o no los mínimos y máximos
    show_min_max = n_clicks % 2 == 1
    return create_figure(show_min_max)

# Exportar como HTML
def save_graph_as_html():
    fig = create_figure(show_min_max=False)  # Crea la figura
    output_path = os.path.join(os.getcwd(), "grafico_desempleo.html")  # Ruta del archivo
    fig.write_html(output_path)  # Exportar como HTML
    print(f"Archivo guardado en: {output_path}")  # Mensaje de confirmación

# Guardar el gráfico antes de ejecutar el servidor
save_graph_as_html()

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
