#### 1.1 Complete los datos de esta tabla. El costo fijo total (CFT) es igual a 175.

```python
import pandas as pd
import numpy as np
import openpyxl

# Definir los datos para cada columna
data = {
    'q': list(range(0, 11)),  # Valores para la columna 'q'
    'IT': [np.nan] * 11,       # Valores vacíos para la columna 'IT'
    'IM': [np.nan] * 11,       # Valores vacíos para la columna 'IM'
    'CFT': [175] * 11,         # Costo Fijo Total constante en todas las filas
    'CVT': [0, 100, 180, 230, 305, 405, 530, 680, 855, 1055, 1285], 
    'CT': [np.nan] * 11,       # Valores vacíos para la columna 'CT'
    'CFMe': [np.nan] * 11,     # Valores vacíos para la columna 'CFMe'
    'CVMe': [np.nan] * 11,     # Valores vacíos para la columna 'CVMe'
    'CM': [np.nan] * 11,       # Valores vacíos para la columna 'CM'
    'BM': [np.nan] * 11,       # Valores vacíos para la columna 'BM'
    'π': [np.nan] * 11         # Valores vacíos para la columna 'π'
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Calcular el Ingreso Total (IT)
df['IT'] = df['q']*150

# Calcular el Ingreso Marginal (IM)
df['IM'] = (df['IT'].diff())/(df['q'].diff())

# Calcular el Costo Total (CT)
df['CT'] = df['CFT'] + df['CVT']

# Calcular el Costo Fijo Medio (CFMe)
df['CFMe'] = round(df['CFT'] / df['q'], 2)

# Calcular el Costo Variable Medio (CVMe)
df['CVMe'] = round(df['CVT'] / df['q'], 2)

# Calcular el Costo Marginal (CM)
df['CM'] = (df['CT'].diff())/(df['q'].diff())

# Calcular el Beneficio Marginal (BM)
df['BM'] = df['IM'] - df['CM']

# Calcular el Beneficio (π)
df['π'] = df['IT'] - df['CT']

print(df)


# Guardar el DataFrame
df.to_excel('Tabla_1.xlsx', index=False)

# Calcular CTMe
df['CTMe'] = df['CT'] / df['q']

# Crear el DataFrame con los datos necesarios
datos = {
    'q': df['q'],
    'CVMe': df['CVMe'],
    'CTMe': df['CTMe'],
    'CM': df['CM'],
    'IM': df['IM']
}

df_datos = pd.DataFrame(datos)

# Guardar el DataFrame en un archivo CSV
df_datos.to_csv('Datos_1_4.csv', index=False)
```

#### 1.2 Para el precio del mercado determine el nivel de producción que maximiza los beneficios y calcule el monto de los beneficios.

```python
# Encontrar el nivel de producción donde BM es cero
nivel_max_beneficio = df[df['BM'] == 0]

# Calcular el monto de los beneficios en ese nivel de producción
beneficio_maximo = nivel_max_beneficio['π'].values

print(f'Nivel de producción que maximiza los beneficios:\n{nivel_max_beneficio}\n')
print(f'Monto de los beneficios en ese nivel de producción: {beneficio_maximo}')
```

#### **1.4** Grafique juntas las curvas de CVMe, CTMe y CM. En esta gráfica represente las curvas de IMg y los respectivos niveles de producción para los precios indicados en las partes 1.2 y 1.3.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
import numpy as np

# Cargar el DataFrame desde el archivo CSV
df = pd.read_csv('Datos_1_4.csv')

# Limpiar los datos eliminando valores NaN e infinitos
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Función para suavizar las líneas
def smooth_line(x, y, num_points=500):
    x_smooth = np.linspace(x.min(), x.max(), num_points)
    spl = make_interp_spline(x, y, k=3)
    y_smooth = spl(x_smooth)
    return x_smooth, y_smooth

# Graficar CVMe con suavizado
x_smooth, y_smooth = smooth_line(df['q'], df['CVMe'])
plt.plot(x_smooth, y_smooth, label='CVMe', color='blue')

# Graficar CTMe con suavizado
x_smooth, y_smooth = smooth_line(df['q'], df['CTMe'])
plt.plot(x_smooth, y_smooth, label='CTMe', color='orange')

# Graficar CM con suavizado
x_smooth, y_smooth = smooth_line(df['q'], df['CM'])
plt.plot(x_smooth, y_smooth, label='CM', color='green')

# Graficar IM con suavizado
x_smooth, y_smooth = smooth_line(df['q'], df['IM'])
plt.plot(x_smooth, y_smooth, label='IM', color='black', linestyle='--')

# Etiquetas y título
plt.xlabel('Nivel de Producción (q)')
plt.ylabel('Precio')
plt.title('Curvas de CVMe, CTMe, CM e IM')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
