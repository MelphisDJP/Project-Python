```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import median, mode, multimode, stdev, pstdev, variance, pvariance

def ordenar_fila(df, indice_fila):
    
    # Seleccionar la fila específica
    fila = df.loc[indice_fila]
    
    # Ordenar la fila de menor a mayor
    fila_ordenada = fila.sort_values(ascending=True)
    
    return fila_ordenada
```

#### CALCULE PRODUCTO TOTAL E INGRESO TOTAL, haciendo la sumatoria de cada producto

```python
df = pd.read_excel('Datos Panaderia.xlsx')

# Calcular Producto Total (sumatoria de cada producto)
producto_total_agua = round(df['Pan de agua'].sum(), 2)
producto_total_mantequilla_jamon = round(df['Pan de mantequilla y jamón'].sum(), 2)
producto_total_ajo = round(df['Pan de ajo'].sum(), 2)

producto_total = producto_total_agua + producto_total_mantequilla_jamon + producto_total_ajo

# Calcular Ingreso Total (sumatoria de ingresos por cada producto)
ingreso_total_agua = round(df['Ingreso x pan agua (RD$)'].sum(), 2)
ingreso_total_mantequilla_jamon = round(df['Ingreso x pan mantequilla y jamón (RD$)'].sum(), 2)
ingreso_total_ajo = round(df['Ingreso x pan de ajo (RD$)'].sum(), 2)

ingreso_total = (ingreso_total_agua + ingreso_total_mantequilla_jamon + ingreso_total_ajo)

# Mostrar resultados
print(f"Producto Total de Pan de Agua: {producto_total_agua}")
print(f"Producto Total de Pan de Mantequilla y Jamón: {producto_total_mantequilla_jamon}")
print(f"Producto Total de Pan de Ajo: {producto_total_ajo}\n")

print(f'El producto Total es: {producto_total:.2f}\n')

print(f"Ingreso Total de Pan de Agua (RD$): {ingreso_total_agua}")
print(f"Ingreso Total de Pan de Mantequilla y Jamón (RD$): {ingreso_total_mantequilla_jamon}")
print(f"Ingreso Total de Pan de Ajo (RD$): {ingreso_total_ajo}\n")

print(f'El Ingreso Total es: ${ingreso_total:.2f}')
```

```python
def variacion_mensual(valores):
    # Formula: Variacion Relativa = (Valor Final - Valor Inicial / Valor Inicial) * 100
    variaciones = []
    for i in range(1, len(valores)):
        valor_inicial = valores[i - 1]
        valor final = valores[i]
        variacion = ((valor_final - valor_inicial)/valor_inicial)*100
        
        variaciones.append(variacion)
        
    return variaciones
        
    
```

#### 5.	Calcule el precio promedio de cada producto para cada año. 

```python
import pandas as pd

# Lee el archivo Excel
df = pd.read_excel('Datos Panaderia.xlsx')

# Convertir la columna 'Meses' a formato datetime si es necesario
df['Meses'] = pd.to_datetime(df['Meses'], format='%B %Y')

df['Año'] = df['Meses'].dt.year

# Filtrar los datos para el año 2020
df_2020 = df[df['Año'] == 2020].copy()

# Filtrar los datos para el año 2021
df_2021 = df[df['Año'] == 2021].copy()

# Calcular el precio por unidad para cada producto en el año 2020
df_2020['Precio Pan de Agua'] = df_2020['Ingreso x pan agua (RD$)'] / df_2020['Pan de agua']
df_2020['Precio Pan de Mantequilla y Jamón'] = df_2020['Ingreso x pan mantequilla y jamón (RD$)'] / df_2020['Pan de mantequilla y jamón']
df_2020['Precio Pan de Ajo'] = df_2020['Ingreso x pan de ajo (RD$)'] / df_2020['Pan de ajo']

# Calcular el precio promedio para cada producto en el año 2020
precio_promedio_pan_agua_2020 = df_2020['Precio Pan de Agua'].mean()
precio_promedio_pan_mantequilla_y_jamon_2020 = df_2020['Precio Pan de Mantequilla y Jamón'].mean()
precio_promedio_pan_ajo_2020 = df_2020['Precio Pan de Ajo'].mean()

# Mostrar el precio promedio para el 2020
print(f'El precio promedio del producto Pan de Agua en el año 2020 es: ${precio_promedio_pan_agua_2020:.2f}')
print(f'El precio promedio del producto Pan de Mantequilla y Jamón en el año 2020 es: ${precio_promedio_pan_mantequilla_y_jamon_2020:.2f}')
print(f'El precio promedio del producto Pan de Ajo en el año 2020 es: ${precio_promedio_pan_ajo_2020:.2f}')

# Calcular el precio por unidad para cada producto en el año 2021
df_2021['Precio Pan de Agua'] = df_2021['Ingreso x pan agua (RD$)'] / df_2021['Pan de agua']
df_2021['Precio Pan de Mantequilla y Jamón'] = df_2021['Ingreso x pan mantequilla y jamón (RD$)'] / df_2021['Pan de mantequilla y jamón']
df_2021['Precio Pan de Ajo'] = df_2021['Ingreso x pan de ajo (RD$)'] / df_2021['Pan de ajo']

# Calcular el precio promedio para cada producto en el año 2021
precio_promedio_pan_agua_2021 = df_2021['Precio Pan de Agua'].mean()
precio_promedio_pan_mantequilla_y_jamon_2021 = df_2021['Precio Pan de Mantequilla y Jamón'].mean()
precio_promedio_pan_ajo_2021 = df_2021['Precio Pan de Ajo'].mean()

print()

# Mostrar el precio promedio para el 2021
print(f'El precio promedio del producto Pan de Agua en el año 2021 es: ${precio_promedio_pan_agua_2021:.2f}')
print(f'El precio promedio del producto Pan de Mantequilla y Jamón en el año 2021 es: ${precio_promedio_pan_mantequilla_y_jamon_2021:.2f}')
print(f'El precio promedio del producto Pan de Ajo en el año 2021 es: ${precio_promedio_pan_ajo_2021:.2f}')

# Guardar los resultados en un archivo CSV
resultados = pd.DataFrame({
    'Producto': ['Pan de Agua', 'Pan de Mantequilla y Jamón', 'Pan de Ajo'],
    'Precio Promedio 2020': [precio_promedio_pan_agua_2020, precio_promedio_pan_mantequilla_y_jamon_2020, precio_promedio_pan_ajo_2020],
    'Precio Promedio 2021': [precio_promedio_pan_agua_2021, precio_promedio_pan_mantequilla_y_jamon_2021, precio_promedio_pan_ajo_2021]
})

resultados.to_csv('resultados_precios.csv', index=False)
```
